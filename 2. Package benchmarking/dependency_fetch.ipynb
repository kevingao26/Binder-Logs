```python
import pandas as pd
import pickle
```

```python
df = pd.read_json("binder-clean.json", orient="index").reset_index()
df
```

```python
dep_list = df["pip"].apply(lambda x: [i.split('==', 1)[0] for i in x]).explode().value_counts().index.to_list()
dep_list
```

```python
with open("dep_list.txt", "wb") as fp:   #Pickling
...   pickle.dump(dep_list, fp, protocol=2)
```

```python
with open("dep_list.txt", "rb") as fp:   # Unpickling
...   b = pickle.load(fp)
```
